### Questões de negócio:
1. A associação ou não com as frases “conversa” de alguma forma com o grau de recomendação para as marcas A e B?
2. Se sim: Cada uma das empresas possui uma verba para gastar em comunicação. A finalidade da companha é melhorar a quantidade de promotores da marca. Em qual/quais atributos cada uma das marcas deve investir?
3. Se não: Somos a Marca A. Temos uma verba para investir em comunicação e queremos trazer pessoas da Marca B? No que devemos investir considerando o que é fortaleza para eles e fraqueza para a gente?

```python
# Importando as bibliotecas
import pandas as pd 

pd.set_option("display.max_columns", None)

import warnings
warnings.filterwarnings("ignore")

# Importando a base de dados
df1 = pd.read_excel('BI_SQL_Alexia.xlsx', skiprows=2)
df2 = pd.read_excel('BI_SQL_Alexia.xlsx', skiprows=2)
```

```python
# Fazendo união das informações da marca A

df1.drop(df1.columns[3], axis=1, inplace=True)

df1.drop(df1.columns[25:], axis=1, inplace=True)

df1.head(2)
```

```python
df1.rename(columns={'Nº respondent.': 'id_participante','Mês da pesquisa':'data','Marca A':'recomendacao'}, inplace=True) 

df1['marca'] = 'Marca A'
df1['marca'].head(2)
```

```python
# Fazendo união das informações da marca B

df2.drop(df2.columns[2], axis=1, inplace=True)

df2.drop(df2.columns[3:25], axis=1, inplace=True)

df2.head(2)
```

```python
df2.rename(columns={'Nº respondent.': 'id_participante','Mês da pesquisa':'data','Marca B':'recomendacao'}, inplace=True) 

df2['marca'] = 'Marca B'
df2['marca'].head(2)
```

```python
for col in df2.columns:
    renomeando = col.replace('.1', '')  
    df2.rename(columns={col: renomeando}, inplace=True)

for col in df1.columns:
    renomeando = col.replace('.1', '')
    df1.rename(columns={col: renomeando}, inplace=True)
```

```python
df = pd.concat([df1, df2], ignore_index=True)
df.head()
```

```python
# Verificando informações da base
print(f"O dataframe possui {df.shape[0]} linhas e {df.shape[1]} colunas.\n")

print('Valores duplicados:', df.duplicated().any().sum())
print('Valores nulos:', df.isna().values.sum())
```

```python
# Criando dados dummy
df.replace({'Yes': 1, 'No': 0}, inplace=True)
```

```python
df.to_excel('dados.xlsx', index=False, sheet_name='Base')
```
